```python
#feature data
X = data.drop('spam',axis=1)
X
```

```python
#target variable
y=data['spam']
y
```

```python
# Standardizing the Feature Vectors
from sklearn.preprocessing import MinMaxScaler
scale = MinMaxScaler()
x_transformed = scale.fit_transform(X)

x_transformed = pd.DataFrame(x_transformed, columns = X.columns)
x_transformed
```

```python
X = x_transformed
X
```

```python
#splitting data into train and test data
x_train,x_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=70)
```

```python
# summarize the dataset
print(X.shape, y.shape)
```

**Tuning Hyperparameters for Perceptron Model**

```python
# create the model
model = Perceptron()
```

```python
# define model evaluation method
cv = RepeatedStratifiedKFold(n_splits=10, n_repeats=5, random_state=30)
# evaluate model
scores = cross_val_score(model, X, y, scoring='accuracy', cv=cv, n_jobs=-1)
# summarize result
print('Mean Accuracy: %.3f (%.3f)' % (mean(scores), std(scores)))
```

```python
# define grid
grid = dict()
grid['eta0'] = [0.0001, 0.001, 0.01, 0.1, 1.0]

# grid search learning rate for the perceptron
# define model
model = Perceptron()
# define model evaluation method
cv = RepeatedStratifiedKFold(n_splits=10, n_repeats=5, random_state=30)
# define grid
grid = dict()
grid['eta0'] = [0.00001, 0.0001, 0.001, 0.01, 0.1, 1.0]
# define search
search = GridSearchCV(model, grid, scoring='accuracy', cv=cv, n_jobs=-1)
# perform the search
results = search.fit(X, y)
# summarize
print('Mean Accuracy: %.3f' % results.best_score_)
print('Config: %s' % results.best_params_)
# summarize all
means = results.cv_results_['mean_test_score']
params = results.cv_results_['params']
for mean, param in zip(means, params):
    print(">%.3f with: %r" % (mean, param))
print()
# Summarizing the results
print('Mean Accuracy: %.3f' % results.best_score_)
```

```python
# define grid
grid = dict()
grid['max_iter'] = [1, 10, 100, 1000, 10000]

# grid search total epochs for the perceptron

# define model
model = Perceptron(eta0=0.0001)
# define model evaluation method
cv = RepeatedStratifiedKFold(n_splits=10, n_repeats=3, random_state=1)
# define grid
grid = dict()
grid['max_iter'] = [1, 10, 100, 1000, 10000]
# define search
search = GridSearchCV(model, grid, scoring='accuracy', cv=cv, n_jobs=-1)
# perform the search
results = search.fit(X, y)
# summarize
print('Mean Accuracy: %.3f' % results.best_score_)
print('Config: %s' % results.best_params_)
# summarize all
means = results.cv_results_['mean_test_score']
params = results.cv_results_['params']
for mean, param in zip(means, params):
    print(">%.3f with: %r" % (mean, param))
print()
# Summarizing the results
print('Mean Accuracy: %.3f' % results.best_score_)
```
